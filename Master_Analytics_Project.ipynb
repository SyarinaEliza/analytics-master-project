```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)

df = pd.DataFrame({
    "date": pd.date_range("2025-01-01", periods=120),
    "sales": np.random.randint(50, 300, 120),
    "calls": np.random.randint(100, 500, 120),
    "stock": np.random.randint(0, 100, 120)
})
```

```python
df["efficiency"] = df["sales"] / df["calls"]
df["stock_gap"] = df["sales"] - df["stock"]

df.head()
```

```python
plt.plot(df["date"], df["sales"])
plt.title("Sales Trend")
plt.xticks(rotation=45)
plt.show()
```

```python
df["mean"] = df["sales"].rolling(7).mean()
df["std"] = df["sales"].rolling(7).std()

df["anomaly"] = abs(df["sales"] - df["mean"]) > 2 * df["std"]

df[df["anomaly"] == True]
```

```python
def insights(df):
    print(" 📊 BUSINESS INSIGHTS")

    if df["anomaly"].sum() > 0:
        print("- Sales anomalies detected")

    if df["stock_gap"].mean() > 50:
        print("- Stock mismatch risk detected")

insights(df)
```
